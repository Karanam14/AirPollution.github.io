```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime 
import matplotlib.dates as md
```

```python
df=pd.read_csv('Pollution.csv')
df.head()
```

```python
df.drop("Entry_id",axis=1,inplace=True)
df.head()
```

```python
#finding out pollution level at BowBazar by hourly update
plt.figure(figsize=(12,8))
plt.rcParams.update({'font.size':15})
x =pd.to_datetime(df['Created_at'])
y=df['BowBazar']
plt.plot_date(x, y)
plt.xlabel("Data & Hour")
plt.ylabel("Pollution Level")
plt.title("Hourly Pollution Level at BowBazar")
```

```python
#finding out pollution level at Gariahat by hourly update
plt.figure(figsize=(12,8))
plt.rcParams.update({'font.size':15})
x =pd.to_datetime(df['Created_at'])
y=df['Gariahat']
plt.plot_date(x, y,color='green')
plt.xlabel("Data & Hour")
plt.ylabel("Pollution Level")
plt.title("Hourly Pollution Level at Gariahat")
```

```python
#finding out pollution level at Esplanede by hourly update
plt.figure(figsize=(12,8))
plt.rcParams.update({'font.size':15})
x =pd.to_datetime(df['Created_at'])
y=df['Esplanede']
plt.plot_date(x, y,color='red')
plt.xlabel("Data & Hour")
plt.ylabel("Pollution Level")
plt.title("Hourly Pollution Level at Esplanede")
```

```python
sns.jointplot(x="BowBazar",y="Gariahat",data=df,color='green')
```
